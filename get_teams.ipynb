```python
import pandas as pd
import numpy as np
import requests
```

```python
teams_url = "https://masseyratings.com/scores.php?s=379387&sub=11590&all=1&mode=3&format=2"

# Get data from the teams_url
teams_data = requests.get(teams_url).text
print(teams_data)

# Convert the teams data to a CSV
teams = pd.read_csv("https://masseyratings.com/scores.php?s=379387&sub=11590&all=1&mode=3&format=2",header=None,index_col=None)
```

```python
teams.rename(columns={0: "id", 1: "team"}, inplace=True)
```

```python
teams.to_csv("2021_teams.csv",index=False)
```
